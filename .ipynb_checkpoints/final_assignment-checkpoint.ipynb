# Feature Engineering

We engineered a set of features designed to reflect real-world patterns in health behaviour, seasonality, and temporal trends. These features were selected to strengthen the predictive capacity of our models while aligning with external factors that influence medication sales.

We began by parsing the Date column into a proper datetime format. From this, we extracted the Year, Month and Week for each entry. These features enable the model to detect long-term trends and month-specific patterns, such as allergy seasons or flu waves, that may recur annually.

To capture the impact of public holidays on medication purchasing behaviour, we cleaned the Holiday column by replacing missing values with None and created a binary column Is_Holiday, which indicates whether the date falls on a holiday. This helps account for unusual spikes in demand around public events, long weekends, or promotional campaigns.

Seasonality was addressed by generating a Season feature based on the month of the year. Months were grouped into four categories: Winter (December to February), Spring (March to May), Summer (June to August), and Fall (September to November).We can visualize broader environmental and behavioural patterns, such as increases in cold and flu medication sales during the colder seasons or allergy relief spikes in spring.

In addition, we defined a custom Is_Flu_Season feature to mark whether a given month falls within the typical flu season. We used calendar months to define flu season as January through April and October through December. This binary indicator (1 for flu season, 0 otherwise) is useful for highlighting time periods when demand for cold, cough, and fever medications tends to rise sharply.

Together, these engineered features provide a richer, more interpretable foundation for visualization and machine learning models. They help capture complex temporal dynamics and public health behaviours that raw data alone would not reveal. By embedding these patterns directly into the feature space, we can improve prediction accuracy and support more meaningful insights from the model outputs.

```python
# Convert date
df['Date'] = pd.to_datetime(df['Date'])

# Basic time features
# Year
df['Year'] = df['Date'].dt.year
# Month
df['Month'] = df['Date'].dt.month
# Weeks in a year
start_date = df['Date'].min()
df['Week'] = ((df['Date'] - start_date).dt.days // 7) + 1

# Holiday column
df['Holiday'] = df['Holiday'].fillna('None')
df['Is_Holiday'] = (df['Holiday'] != 'None').astype(int)

# Season from month
def get_season(month):
    if month in [12, 1, 2]: return 'Winter'
    elif month in [3, 4, 5]: return 'Spring'
    elif month in [6, 7, 8]: return 'Summer'
    else: return 'Fall'
df['Season'] = df['Month'].apply(get_season)

# Flu season from month: OctDec and JanApr
df['Is_Flu_Season'] = df['Month'].apply(lambda x: 1 if (x >= 10 or x <= 4) else 0)
```

# Train machine learning models and evaluate

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor, AdaBoostRegressor
from sklearn.neighbors import KNeighborsRegressor
from sklearn.tree import DecisionTreeRegressor, plot_tree
from xgboost import XGBRegressor
from sklearn.metrics import mean_squared_error, r2_score
```

The data was prepped for the machine learning models by selecting the most relevant factors that could influence medication sales. These factors include type of medication, the region where it is sold, the season, marketing spending, Google search trends, flu case numbers, and environmental conditions such as temperature, humidity, and pollen count. It also includes time-based indicators like the week of the year and whether it's during flu season or a holiday. Since some of these variables are written as categories (like region or season), they were converted into numerical values so that the machine learning model can understand and process them. 

```python
# Select relevant features and encode categorical ones
features = ['Medication', 'Region', 'Season', 'Google_Trends', 'Marketing_Spend', 
            'Temperature', 'Flu_Cases', 'Humidity', 'Week', 'Is_Flu_Season', 
            'Is_Holiday', 'Pollen_Count']

df_model = df.copy()
le_med = LabelEncoder()
le_reg = LabelEncoder()
le_season = LabelEncoder()
df_model['Medication'] = le_med.fit_transform(df_model['Medication'])
df_model['Region'] = le_reg.fit_transform(df_model['Region'])
df_model['Season'] = le_season.fit_transform(df_model['Season'])

X = df_model[features]
y = df_model['Sales']

```

Next, all the input features were scaled so that the machine learning models can interpret them correctly. Many models, especially those based on distance or gradient calculations, are sensitive to the range of values in the data. For example, marketing spend might range in the thousands, while something like humidity is typically between 0 and 100. If left unscaled, features with larger ranges could dominate the learning process. Stnadardizing will transforms each feature so that it has a mean of 0 and a standard deviation of 1. This helps the model treat all features more equally and often leads to better and more stable performance.

```python
# Scale the data
scaler = StandardScaler()
X_scaled = pd.DataFrame(scaler.fit_transform(X), columns=features)
```

Next, the dataset was divided into two separate parts: one for training the machine learning model and one for testing how well it performs. Specifically, 80% of the data is used to train the model (X_train, y_train), and the remaining 20% is set aside to evaluate its accuracy on new, unseen data (X_test, y_test). This helps ensure that the model doesnt just memorize the training data, but can generalize its predictions to other real-world situations. The random_state=42 ensures that the split is reproducible.

```python
# Split the balanced data
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)
```

Six different machine learning models were created. Their performance on both the training and testing datasets. The goal was to see how well each model can learn from the data and generalize to new, unseen information and predict medication sales.

Each model was trained using the training data (X_train, y_train). After training, the model made predictions on both the training set (to check how well it has learned) and the test set (to see how well it performs on unseen data).

Two evaluation metrics are calculated for each:
1. MSE (Mean Squared Error): Measures how far off the predictions are from the actual sales values  lower is better.
2. R (R-squared Score): Measures how much of the variability in sales the model explains  closer to 1 is better.



```python
# Initialize models
models = {
    "Linear Regression": LinearRegression(),
    "Random Forest": RandomForestRegressor(n_estimators=70, random_state=10),
    "AdaBoost": AdaBoostRegressor(n_estimators=80, random_state=10),
    "KNN": KNeighborsRegressor(),
    "Decision Tree": DecisionTreeRegressor(max_depth=7),
    "XGBoost": XGBRegressor(n_estimators=80, random_state=10)
}

# Evaluate both train and test performance for each model
train_test_results = {}

for name, model in models.items():
    model.fit(X_train, y_train)
    preds_test = model.predict(X_test)
    preds_train = model.predict(X_train)
    
    mse_test = mean_squared_error(y_test, preds_test)
    r2_test = r2_score(y_test, preds_test)
    mse_train = mean_squared_error(y_train, preds_train)
    r2_train = r2_score(y_train, preds_train)
    
    train_test_results[name] = {
        'Train MSE': round(mse_train, 2),
        'Train R2': round(r2_train, 4),
        'Test MSE': round(mse_test, 2),
        'Test R2': round(r2_test, 4)
    }

# Create and show DataFrame
train_test_df = pd.DataFrame(train_test_results).T
train_test_df
```

Random Forest and XGBoost performed the best overall, achieving high R scores on both train and test sets, meaning they accurately captured patterns in the data without overfitting. Linear Regression had moderate performance, suggesting it can model some trends, but is limited by its simplicity. AdaBoost performed very poorly, with almost no predictive power, likely due to how it handles noise or outliers. KNN and Decision Tree had decent results but slightly lower test performance than Random Forest or XGBoost, which might suggest some overfitting or limited generalization. The resulting decision tree was plotted below.

```python
# Plot Decision Tree
plt.figure(figsize=(20, 10))
plot_tree(models["Decision Tree"], feature_names=X.columns, filled=True, rounded=True, fontsize=10)
plt.title("Decision Tree Visualization")
plt.tight_layout()
plt.show()
```

We performed hyperparameter tuning to improve the accuracy and generalization of our machine learning models. Since different models can behave very differently depending on their configuration, we used GridSearchCV to systematically explore a range of parameter combinations and identify the optimal settings for each model.

For each model, we defined a set of hyperparameters to test. For example, for the Random Forest, we varied the number of trees (n_estimators), maximum tree depth (max_depth), and the minimum number of samples required to split a node (min_samples_split). Similar tuning ranges were created for AdaBoost, KNN, Decision Tree, and XGBoost, focusing on parameters most likely to influence model performance.

We then applied GridSearchCV with 3-fold cross-validation to evaluate all combinations of parameters using the R score as the performance metric. For each model, the grid search returned the best set of hyperparameters based on cross-validated performance. We retrained each model using these optimal settings and assessed their predictive ability on the test set by calculating the Mean Squared Error (MSE) and R score.

The tuned performance metrics were compiled into a summary table, showing the best parameters for each model along with their corresponding test results. This process helped us refine our models and confirmed that Random Forest and XGBoost continued to outperform the others even after tuning.

```python
from sklearn.model_selection import GridSearchCV

# Define hyperparameter grids for tuning
param_grids = {
    "Random Forest": {
        'n_estimators': [50, 100, 150],
        'max_depth': [5, 10, 20, None],
        'min_samples_split': [2, 5]
    },
    "AdaBoost": {
        'n_estimators': [50, 100, 200],
        'learning_rate': [0.01, 0.1, 1]
    },
    "KNN": {
        'n_neighbors': [3, 5, 7, 9],
        'weights': ['uniform', 'distance']
    },
    "Decision Tree": {
        'max_depth': [3, 5, 10, None],
        'min_samples_split': [2, 5, 10]
    },
    "XGBoost": {
        'n_estimators': [50, 100, 150],
        'max_depth': [3, 5, 7],
        'learning_rate': [0.01, 0.1, 0.2]
    }
}

# Models to tune
base_models = {
    "Random Forest": RandomForestRegressor(random_state=10),
    "AdaBoost": AdaBoostRegressor(random_state=10),
    "KNN": KNeighborsRegressor(),
    "Decision Tree": DecisionTreeRegressor(random_state=10),
    "XGBoost": XGBRegressor(random_state=10)
}

# Store tuned results
tuned_results = {}

# Tune models using GridSearchCV
for name, model in base_models.items():
    grid = GridSearchCV(model, param_grids[name], cv=3, scoring='r2', n_jobs=-1)
    grid.fit(X_train, y_train)
    best_model = grid.best_estimator_
    preds = best_model.predict(X_test)
    mse = mean_squared_error(y_test, preds)
    r2 = r2_score(y_test, preds)
    tuned_results[name] = {
        'Best Params': grid.best_params_,
        'Tuned Test MSE': round(mse, 2),
        'Tuned Test R2': round(r2, 4)
    }

# Display results
tuned_results_df = pd.DataFrame(tuned_results).T
tuned_results_df
```

After evaluating several models, we decided to move forward with Random Forest for further analysis due to its strong and consistent performance. To maximize its predictive accuracy, we conducted a focused hyperparameter tuning process using GridSearchCV. We tested different configurations for the number of trees (n_estimators), tree depth (max_depth), and the minimum number of samples required to split a node (min_samples_split) using 3-fold cross-validation and R as the scoring metric.

Once the optimal parameters were identified, we retrained the model and evaluated it on the test dataset. We then calculated the models predictive performance using Mean Squared Error (MSE) and R score, which confirmed the model's ability to generalize well to unseen data.

Finally, we extracted the feature importance scores from the tuned model. These scores indicate which variables had the most influence on predicting medication sales. Features such as marketing spend, Google search trends, region, and flu case counts were among the top contributors, offering valuable insight into the key drivers of medication demand. This analysis not only improved the models performance but also helped highlight the most impactful predictors for forecasting purposes.

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import GridSearchCV

# Define the hyperparameter grid for Random Forest
rf_param_grid = {
    'n_estimators': [50, 100, 150],
    'max_depth': [5, 10, 20, None],
    'min_samples_split': [2, 5]
}

# Instantiate and run GridSearchCV for Random Forest only
rf_model = RandomForestRegressor(random_state=10)
rf_grid = GridSearchCV(rf_model, rf_param_grid, cv=3, scoring='r2', n_jobs=-1)
rf_grid.fit(X_train, y_train)

# Best model
best_rf = rf_grid.best_estimator_

# Predict and evaluate on test data
rf_preds = best_rf.predict(X_test)
rf_mse = mean_squared_error(y_test, rf_preds)
rf_r2 = r2_score(y_test, rf_preds)

# Extract feature importance
rf_feature_importance = pd.DataFrame({
    'Feature': X_train.columns,
    'Importance': best_rf.feature_importances_
}).sort_values(by='Importance', ascending=False)

rf_feature_importance

```

```python
# Save results for summary
rf_summary = {
    "Best Params": rf_grid.best_params_,
    "Test MSE": round(rf_mse, 2),
    "Test R2": round(rf_r2, 4)
}

rf_summary
```

And finally, we summarized the performance of the tuned Random Forest model. After identifying the optimal configuration, and a minimum sample split of 5, we recorded its test performance. The model achieved a Mean Squared Error (MSE) of 51,717.82, and an R score of 0.7621 on the test dataset. This means the model was able to explain approximately 76% of the variability in medication sales, confirming its strong ability to generalize to new data. These results validated our decision to use Random Forest as the primary model for predicting medication demand.
